```python
import os

import matplotlib.pyplot as plt
import numpy as np
try:
    import open3d as o3d
except ImportError:
    import sys
    print(sys.exc_info())
import pandas as pd

from dosipy.utils.viz import scatter_3d
from utils import load_data
```

```python
antenna = 'DipoleVertical'
distance = 5

path = os.path.join('data', 'emf_surface_distribution')
    
ExRe_df = pd.read_csv(
    os.path.join(path, f'{antenna}_Re_Ex_d{distance}mm.txt'),
    sep='\s+', comment='%', header=None, usecols=[0, 1, 2], names=['x', 'y', 'z']
)
EyIm_df = pd.read_csv(
    os.path.join(path, f'{antenna}_Im_Ey_d{distance}mm.txt'),
    sep='\s+', comment='%', header=None, usecols=[0, 1, 2], names=['x', 'y', 'z']
)

for idx, (row1, row2) in enumerate(zip(ExRe_df.values[:-1], EyIm_df.values)):
    thresh = sum(row1 - row2)
    if thresh != 0.0:
        print(idx)
        break
```
